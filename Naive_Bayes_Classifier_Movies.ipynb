```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Customer-Churn.csv')
df.head()
```

```python
df.tail()
```

```python
df.groupby(df.gender).describe()
```

```python
df['Detect'] = df.gender.apply(lambda x: 1 if x == 'Male' else 0)
```

```python
df.head()
```

```python
#Train test & split
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.StreamingMovies, df.Detect, test_size = 0.20)
```

```python
#Covert Streaming Movies to numbers
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v = CountVectorizer()
```

```python
x_train_count = v.fit_transform(x_train.values)
x_train_count.toarray()[:3]
```

```python
#Use naive bayes Multinomial for prediction
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
model.fit(x_train_count, y_train)
```

```python
#make the prediction
movies = [
    'Yes',
    'No',
    'No',
    'Yes'
]

movies_count = v.transform(movies)
model.predict(movies_count)
```

```python
#Find the testing score
x_test_count = v.transform(x_test)
model.score(x_test_count, y_test)
```
