Predict if the salary of a person is more than 100 000.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('salaries.csv')
df
```

```python
#I dentify the target
target = df.salary_more_then_100k
target
```

```python
#Drop the column that you identifyed as target
df = df.drop('salary_more_then_100k', axis = 'columns')
df.head()
```

```python
#Contvert the string to numbers from the columns
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
df.company = label_encoder.fit_transform(df.company)
df.job = label_encoder.fit_transform(df.job)
df.degree = label_encoder.fit_transform(df.degree)
df.head()
```

```python
#Train the model
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
#Fit the model
model.fit(df, target)
```

```python
#Make the the prediction for company: 2, job: 2, degree: 1 and company: 2, job: 1, degree: 0
toPredict = np.array([[2, 2, 1]])
predictedSalary = model.predict(toPredict)
```

```python
predictedSalary
```

```python
target
```

```python
model.score(df, target)
```
