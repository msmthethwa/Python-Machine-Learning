```python
import pandas as pd
import numpy as np
```

```python
#Load the dataset
df = pd.read_csv('spam.csv')
```

```python
df.head()
```

```python
# Group by 'Category' and display descriptive statistics
df.groupby('Category').describe()
```

```python
# Create a new column 'Detect' where 'spam' is 1 and 'ham' is 0
df['Detect'] = df.Category.apply(lambda x:1 if x == 'spam' else 0)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Split the dataset into training and testing sets (80% train, 20% test)
x_train, x_test, y_train, y_test = train_test_split(df.Message, df.Detect, test_size = 0.20)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Initialize CountVectorizer to convert text to a matrix of token counts
v = CountVectorizer()
```

```python
# Fit and transform the training data
x_train_count = v.fit_transform(x_train.values)

# Display the first 3 rows of the training data as an array
x_train_count.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
# Initialize the Multinomial Naive Bayes model
model = MultinomialNB()
```

```python
# Train the model with the training data
model.fit(x_train_count, y_train)
```

```python
# Example emails for prediction
emails = [
    'I‘m going to try for 2 months ha ha only joking',
    'As a valued customer, I am pleased to advise you that following recent review of your Mob No. you are awarded with a £1500 Bonus Prize, call 09066364589',
    'Hey mohan, can we get together to watch football game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]

# Transform the example emails to count matrix
emails_count = v.transform(emails)

# Predict whether the example emails are spam or not
model.predict(emails_count)
```

```python
# Transform the test data to count matrix
x_test_count = v.transform(x_test)

# Evaluate the model's accuracy on the test data
model.score(x_test_count, y_test)
```
