Given those home prices find out price of homes whose areas is: 33000 square feet; 5000 square feet.

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
#Load the dataset
df = pd.read_csv('homeprices.csv')
df
```

```python
#Prepare the dataset
x = df[['area']] #feature
y = df['price'] #target
```

```python
#Create linear regression model
model = LinearRegression()
```

```python
#Fit the model
model.fit(x,y)
```

```python
areaToPredict = np.array([[3300]])
price = model.predict(areaToPredict)
```

```python
price
```

```python
model.coef_
```

```python
model.intercept_
```
