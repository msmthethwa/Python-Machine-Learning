Linear Regression: Predicting housing prices utilizing USA_Housing.csv

Avg. Area Income: 53298.19 | 
Avg. Area House Age:4 | 
Avg. Area Number of Rooms: 5 | 
Avg. Area Number of Bedrooms: 3 | 
Area Population: 56172  | 
Address: 8796 Joseph Burg

```python
import pandas as pd
import numpy as np
```

```python
#Load the dataset and print the first five rows
df = pd.read_csv('USA_Housing.csv')
df.head()
```

```python
df['Avg. Area Income'] = df['Avg. Area Income'].round(2)
df['Avg. Area House Age'] = df['Avg. Area House Age'].astype(int)
df['Avg. Area Number of Rooms'] = df['Avg. Area Number of Rooms'].astype(int)
df['Avg. Area Number of Bedrooms'] = df['Avg. Area Number of Bedrooms'].astype(int)
df['Area Population'] = df['Area Population'].astype(int)
df['Price'] = df['Price'].apply(lambda x: '{:.2f}'.format(x))
df.head()
```

```python
#Handle string for address
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
df['Address'] = label_encoder.fit_transform(df.Address)
df.head()
```

```python
#Prepare the data
x = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms', 'Avg. Area Number of Bedrooms', 'Area Population', 'Address']]
y = df['Price']
```

```python
#Create a linear regression model
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
#Fit the model
model.fit(x,y)
```

```python
'''Make the prediction for 
Avg. Area Income: 53298.19 | Avg. Area House Age:4 | Avg. Area Number of Rooms: 5 | Avg. Area Number of Bedrooms: 3 | Area Population: 56172 | 
Address: 8796 Joseph Burg'''
prediction = np.array([[53298.19, 4, 5, 3, 56172,4069]])
predictedPrice = model.predict(prediction)
```
